```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('income.csv')
df
```

```python
df=df.rename(columns={'income($)':'salary','count':'counts'})
df
```

```python
sns.set(rc={'figure.figsize':(11.7,8.7)})
g=sns.barplot(x='salary',y='counts',data=df)
g.set_xticklabels(g.get_xticklabels(),rotation=45,horizontalalignment='right')
```

```python
sns.set(rc={'figure.figsize':(11.7,8.7)})
g=sns.barplot(x='salary',y='counts',data=df)
g.set_xticklabels(g.get_xticklabels(),rotation=45,horizontalalignment='right')
# plt.xscale('log')
g.set(xscale='log')
```
