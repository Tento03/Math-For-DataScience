```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df=pd.read_csv('heights.csv')
df
```

```python
Z_Score=(df['height'] - df['height'].mean())/df['height'].std()
Z_Score
```

Z Score adalah seberapa jauh standard deviasi dari mean dataset

```python
df['Z-Score']=Z_Score
df
```

Ini berarti pada nilai pertama,yaitu Z-Score dengan 1.94 berarti std nya sejauh 1.94 dari mean dataset

```python
mean_per_height=df['height'].mean()
mean_per_height
```

```python
std_per_height=df['height'].std()
std_per_height
```

```python
std_dataset=(df['height'][0] - mean_per_height)/std_per_height
std_dataset
```
